```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import librosa
import librosa.display
import soundfile
from IPython.display import display, Audio
```

```python
path = 'violin.flac'
data, sr = librosa.load(path, sr=44100, mono=True)
display(Audio(path))
```

```python
def separate_array(array, func):
    chunks = []
    prev_val = array[0]
    current_chunk = [prev_val]

    for i in range(1, len(array)):
        val = array[i]
        if (func(val, prev_val)):
            chunks.append(current_chunk)
            current_chunk = [] 
        current_chunk.append(val)
        prev_val = val

    chunks.append(current_chunk)
    return chunks
```

```python
separate_array([1, 2, 3, 2, 1, 3, 4], lambda x, p: x < p)
```

```python
def separate_wavesets(array, min_length):
    chunks = np.empty(0)
    prev_val = np.array(0.)
    current_chunk = np.array(prev_val)
    chunk_length = 1

    for i in range(1, len(array)):
        val = array[i]
        if (prev_val < 0. and val >= 0 and chunk_length >= min_length):
            chunks = np.append(chunks, current_chunk)
            current_chunk = np.empty(0)
            chunk_length = 0
        current_chunk = np.append(current_chunk, np.array(val))
        chunk_length = chunk_length + 1
        prev_val = val

    chunks = np.append(chunks, [current_chunk])
    return chunks


wavesets = separate_wavesets(data, 1024)

wavesets
```
